```python
#-------------------------------------------
# A One-dimensional Euler Method
#-------------------------------------------
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# y' = y-(1/2)t^2+1, 0<= t <= 2, y(0) = 0.5

# Range of t 0.0 <= t <= 2.0
t_0 = 0.0
t_1 = 2.0

Interval = 10
dt = (t_1-t_0)/float(Interval)  #size 

t = np.arange( t_0, t_1+dt, dt ) #Container for t
y = np.zeros(t.size) #Container for y

#Initialize value
t[0] = 0.0
y[0] = 0.5

#ODE
def Func_ODE( t, y ):
    return y - 0.5*t**2 + 1 
 
# Euler method Implementation
for i in range(1,Interval+1):
   y[i] = y[i-1] + dt*Func_ODE(t[i-1], y[i-1])

plt.plot( t, y)
plt.xlabel('t') 
plt.ylabel('y')
plt.title( "One dimensional ODE Euler Method")
```

```python
#-------------------------------------------
# A One-dimensional ODE Modified Euler Method
#-------------------------------------------
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# y' = y-(1/2)t^2+1, 0<= t <= 2, y(0) = 0.5

# Range of t 0.0 <= t <= 2.0
t_0 = 0.0
t_1 = 2.0

Interval = 10
dt = (t_1-t_0)/float(Interval)  #size 

t = np.arange( t_0, t_1+dt, dt ) #Container for t
y = np.zeros(t.size) #Container for y
 
t[0] = 0.0
y[0] = 0.5

#ODE
def Func_ODE( t, y ):
    return y - 0.5*t**2 + 1 
 
# Modified Euler's method Implementation
for i in range(1,Interval+1):
    Func1 = Func_ODE( t[i-1], y[i-1] )
    Func2 = Func_ODE( t[i],y[i-1] + dt * Func1 )
    y[i] =y[i-1] + dt * ( Func1 + Func2 )/2.0
     
plt.plot( t, y)
plt.xlabel('t') 
plt.ylabel('y')
plt.title( "One dimensional ODE Modified Euler Method")
```

```python
#-------------------------------------------
# A One-dimensional ODE Runge-Kutta method
#-------------------------------------------
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# y' = y-(1/2)t^2+1, 0<= t <= 2, y(0) = 0.5

# Range of t 0.0 <= t <= 2.0
t_0 = 0.0
t_1 = 2.0

Interval = 10
dt = (t_1-t_0)/float(Interval)  #size 
 
t = np.arange( t_0, t_1+dt, dt ) #Container for t
y = np.zeros(t.size) #Container for y
 
t[0] = 0.0
y[0] = 0.5

#ODE
def Func_ODE( t, y ):
    return y - 0.5*t**2 + 1 
 
# Runge-Kutta method Implementation
for i in range(1,Interval+1):   
        k1 = dt * Func_ODE( t[i-1], y[i-1] )
        k2 = dt * Func_ODE( t[i-1] + dt / 2.0, y[i-1] + k1 / 2.0 )
        k3 = dt * Func_ODE( t[i-1] + dt / 2.0, y[i-1] + k2 / 2.0 )
        k4 = dt * Func_ODE( t[i], y[i-1] + k3 )
        y[i] = y[i-1] +  k1/6 + k2/3 + k3/3 + k4/6  

plt.plot( t, y )
plt.xlabel('t')
plt.ylabel('y') 
plt.title( "One dimensional ODE Runge-Kutta Method")
```

```python
#-------------------------------------------
#A system of ODEs Euler method
#-------------------------------------------
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# y' = 2x+y, x' = x + y, x(0) = 4, y(0)=3

# Range of t 0.0 <= t <= 2.0
t_0 = 0.0
t_1 = 2.0

Interval = 10
dt = (t_1-t_0)/float(Interval)  #size 

t = np.arange( t_0, t_1+dt, dt ) #Container for t
x = np.arange(t.size)#Container for x
y = np.zeros(t.size) #Container for y

#Initialize value
t[0] = 0.0
x[0] = 4
y[0] = 3

#ODE
def Func_fx( x, t ):
    return 2*x + t 

def Func_fy( y, t ):
    return t+y
 
# Euler method Implementation
for i in range(1,Interval+1):
    x[i] = x[i-1] + dt*Func_fx(t[i-1], x[i-1])
    y[i] = y[i-1] + dt*Func_fy(t[i-1], y[i-1])
 
plt.plot( t, x, 'g', label = 'x(t)')
plt.plot( t, y, 'b', label = 'y(t)')
legend = plt.legend(loc = 'upper left',shadow = True, fontsize = 'x-large')
plt.title( "A system of ODE Euler Method")
```

```python
#-------------------------------------------
#A system of ODEs Modified Euler Method
#-------------------------------------------
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# y' = 2x+y, x' = x + y, x(0) = 4, y(0)=3

# Range of t 0.0 <= t <= 2.0
t_0 = 0.0
t_1 = 2.0

Interval = 10
dt = (t_1-t_0)/float(Interval)  #size 

t = np.arange( t_0, t_1+dt, dt ) #Container for t
x = np.arange(t.size)#Container for x
y = np.zeros(t.size) #Container for y

#Initialize value
t[0] = 0.0
x[0] = 4
y[0] = 3

#ODE
def Func_fx( x, t ):
    return 2*x + t 

def Func_fy( y, t ):
    return t+y
 
# Modified Euler method Implementation
for i in range(1,Interval+1):
    Func_x1 = Func_fx( t[i-1], x[i-1] )
    Func_x2 = Func_fx( t[i],x[i-1] + dt * Func_x1 )
    x[i] =x[i-1] + dt * ( Func_x1 + Func_x2 )/2.0
    
    Func_y1 = Func_fy( t[i-1], y[i-1] )
    Func_y2 = Func_fy( t[i],y[i-1] + dt * Func_y1 )
    y[i] =y[i-1] + dt * ( Func_y1 + Func_y2 )/2.0
    
plt.plot( t, x, 'g', label = 'x(t)')
plt.plot( t, y, 'b', label = 'y(t)')
legend = plt.legend(loc = 'upper left',shadow = True, fontsize = 'x-large')
plt.title( "A system of ODE Modified Euler Method")
```

```python
#-------------------------------------------
#A system of ODEs Runge-Kutta Method
#-------------------------------------------
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# y' = 2x+y, x' = x + y, x(0) = 1, y(0)=3

# Range of t 0.0 <= t <= 2.0
t_0 = 0.0
t_1 = 2.0

Interval = 10
dt = (t_1-t_0)/float(Interval)  #size 

t = np.arange( t_0, t_1+dt, dt ) #Container for t
x = np.arange(t.size)#Container for x
y = np.zeros(t.size) #Container for y

#Initialize value
t[0] = 0.0
x[0] = 4
y[0] = 3

#ODE
def Func_fx( x, t ):
    return 2*x + t 

def Func_fy( y, t ):
    return t+y
 
# Modified Euler method Implementation
for i in range(1,Interval+1):
        k1_x = dt * Func_fx( t[i-1], x[i-1] )
        k1_y = dt * Func_fy( t[i-1], y[i-1] )
        
        k2_x = dt * Func_fx( t[i-1] + dt / 2.0, x[i-1] + k1_x / 2.0 )
        k2_y = dt * Func_fy( t[i-1] + dt / 2.0, y[i-1] + k1_y / 2.0 )
        
        k3_x = dt * Func_fx( t[i-1] + dt / 2.0, x[i-1] + k2_x / 2.0 )
        k3_y = dt * Func_fy( t[i-1] + dt / 2.0, y[i-1] + k2_y / 2.0 )
        
        k4_x = dt * Func_fx( t[i], x[i-1] + k3_x )
        k4_y = dt * Func_fy( t[i], y[i-1] + k3_y )
        
        x[i] = x[i-1] +  k1_x/6 + k2_x/3 + k3_x/3 + k4_x/6  
        y[i] = y[i-1] +  k1_y/6 + k2_y/3 + k3_y/3 + k4_y/6  
    
plt.plot( t, x, 'g', label = 'x(t)')
plt.plot( t, y, 'b', label = 'y(t)')
legend = plt.legend(loc = 'upper left',shadow = True, fontsize = 'x-large')
plt.title( "A system of ODE Runge-Kutta Method")
```

```python
#-------------------------------------------
# A second order ODE Euler Method
#-------------------------------------------
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#y'' + 2y' +y = 0, y(0) = 1, y'(0)=1

# Range of t 0.0 <= t <= 2.0
t_0 = 0.0; 
t_1 = 2.0;

Interval = 10

dt = (t_1-t_0)/float(Interval) # size

t = np.arange(t_0, t_1+dt, dt) #Container for t
y = np.zeros(t.size) #container for y
z = np.zeros(t.size) #container for z

#Initialize value
t[0] = 0.0
y[0] = 1
z[0] = 1

# Function to return first order ode
def Func_ODE(y,z):
    #dy/dt
    dy = z;
    #dz/dt
    dz = y - 2*z;
    return dy,dz

# Function to map y
def Func_fy(y,z):
    [dy,dz] = Func_ODE(y,z)
    return dy

# Function to map z
def Func_fz(y,z):
    [dy,dz] = Func_ODE(y,z)
    return dz  

#Implementation of Euler method
for i in range(1,Interval+1):  
    y[i] = y[i-1] + dt*Func_fy(y[i-1],z[i-1])
    z[i] = z[i-1] + dt*Func_fz(y[i-1],z[i-1])

plt.plot(t,y)
plt.xlabel('t')
plt.ylabel('y')
plt.title('A second order ODE Euler Method')
```

```python
#-------------------------------------------
# A second order ODE Runge-Kutta Method
#-------------------------------------------
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#y'' + 2y' +y = 0, y(0) = 1, y'(0)=1

# Range of t 0.0 <= t <= 2.0
t_0 = 0.0; 
t_1 = 2.0;

Interval = 10

dt = (t_1-t_0)/float(Interval) # size

t = np.arange(t_0, t_1+dt, dt) #Container for t
y = np.zeros(t.size) #container for y
z = np.zeros(t.size) #container for z

#Initialize value
t[0] = 0.0
y[0] = 1
z[0] = 1

# Function to return first order ode
def Func_ODE(y,z):
    #dy/dt
    dy = z;
    #dz/dt
    dz = y - 2*z;
    return dy,dz

# Function to map y
def Func_fy(y,z):
    [dy,dz] = Func_ODE(y,z)
    return dy

# Function to map z
def Func_fz(y,z):
    [dy,dz] = Func_ODE(y,z)
    return dz  

#Implementation of Runge-Kutta method
for i in range(1,Interval+1):     
    k1_y = dt*Func_fy(y[i-1],z[i-1])
    k1_z = dt*Func_fz(y[i-1],z[i-1])
    
    k2_y = dt*Func_fy(y[i-1]+dt/2,z[i-1]+k1_z/2)
    k2_z = dt*Func_fz(y[i-1]+dt/2,z[i-1]+k1_z/2)

    k3_y = dt*Func_fy(y[i-1]+dt/2, z[i-1]+k2_z/2)
    k3_z = dt*Func_fz(y[i-1]+dt/2, z[i-1]+k2_z/2)

    k4_y = dt*Func_fy(y[i],z[i-1]+k3_z)
    k4_z = dt*Func_fz(y[i],z[i-1]+k3_z)

    y[i] = y[i-1] + k1_y/6 + k2_y/3 + k3_y/3 + k4_y/6
    z[i] = z[i-1] + k1_z/6 + k2_z/3 + k3_z/3 + k4_z/6
    
plt.plot(t,y)
plt.xlabel('t')
plt.ylabel('y')
plt.title('A second order ODE Runge-Kutta Method')
```
